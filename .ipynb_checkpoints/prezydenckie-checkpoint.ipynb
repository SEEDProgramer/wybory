```python
debile = df.keys().tolist()[33:]
```

```python
df_gmina = df.groupby(['Gmina']).sum()
```

```python
df_gmina.nlargest(3,'Liczba głosów ważnych oddanych łącznie na wszystkich kandydatów')
```

```python
df_deb=df_gmina.nlargest(3,'Liczba głosów ważnych oddanych łącznie na wszystkich kandydatów')
for a in debile:
    print(df_gmina.nlargest(3,a))
    
```
